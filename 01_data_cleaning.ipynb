```python
import pandas as pd

# Load dataset
url = "https://raw.githubusercontent.com/owid/co2-data/master/owid-co2-data.csv"
df = pd.read_csv(url)

# Info about columns and missing data
print("Column types and missing data:")
df.info()

# Statistical summary
print("\nStatistical summary:")
display(df.describe())

# All column names
print("\nAll column names:")
print(df.columns.tolist())

# Preview first few rows
print("\nPreview first 5 rows:")
display(df.head())
```

```python
ag_df = df[df['country'] == 'Afghanistan'] # Afghanistan Data Frame
display(ag_df.head())
display(ag_df.tail())
```

```python
columns_to_keep = [
    "country", "year", "population", "gdp", "co2", "co2_per_capita",
    "coal_co2", "oil_co2", "gas_co2", "cement_co2"
]

# Keep only existing columns
columns_to_keep_existing = [col for col in columns_to_keep if col in df.columns]
df = df[columns_to_keep_existing]

# Calculate gdp_per_capita
df['gdp_per_capita'] = df['gdp'] / df['population']

# Check missing values
print(df.isnull().sum())

# Fill CO2 per capita missing values with 0 (or drop if needed)
df['co2_per_capita'].fillna(0, inplace=True)

# Drop rows without country or year
df.dropna(subset=['country', 'year'], inplace=True)
```

```python
# Sort first to ensure correct chronological order
df = df.sort_values(by=['country', 'year'])

# Fill forward for population and gdp (per country)
df['population'] = df.groupby('country')['population'].ffill()
df['gdp'] = df.groupby('country')['gdp'].ffill()

# Compute gdp_per_capita (new column)
df['gdp_per_capita'] = df['gdp'] / df['population']

# Verify remaining missing values for main metrics
df[['population', 'gdp', 'co2', 'co2_per_capita', 'gdp_per_capita']].isnull().sum()
```

```python
# Keep only rows where GDP and CO2 data are available
df_clean = df.dropna(subset=['gdp', 'co2'])

# Reset index
df_clean = df_clean.reset_index(drop=True)

# Confirm cleaning
df_clean.info()
```

```python
import os

# Define the output directory path
output_dir = "/content/data/cleaned"

# Create the directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Define the full output path
output_path = os.path.join(output_dir, "owid_co2_cleaned.csv")

# Save cleaned dataset to a CSV file
df_clean.to_csv(output_path, index=False)

print(f"✅ Cleaned dataset saved to: {output_path}")
```
