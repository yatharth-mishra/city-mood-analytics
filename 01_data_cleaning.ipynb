```python
import pandas as pd

# Load dataset
url = "https://raw.githubusercontent.com/owid/co2-data/master/owid-co2-data.csv"
df = pd.read_csv(url)

# Info about columns and missing data
print("Column types and missing data:")
df.info()

# Statistical summary
print("\nStatistical summary:")
display(df.describe())

# All column names
print("\nAll column names:")
print(df.columns.tolist())

# Preview first few rows
print("\nPreview first 5 rows:")
display(df.head())
```

```python
ag_df = df[df['country'] == 'Afghanistan'] # Afghanistan Data Frame
display(ag_df.head())
display(ag_df.tail())
```

```python
columns_to_keep = [
    "country", "year", "population", "gdp", "co2", "co2_per_capita",
    "coal_co2", "oil_co2", "gas_co2", "cement_co2"
]

# Keep only existing columns
columns_to_keep_existing = [col for col in columns_to_keep if col in df.columns]
df = df[columns_to_keep_existing]

# Calculate gdp_per_capita
df['gdp_per_capita'] = df['gdp'] / df['population']
```

```python
# Check missing values
print(df.isnull().sum())

# Fill CO2 per capita missing values with 0 (or drop if needed)
df['co2_per_capita'].fillna(0, inplace=True)

# Drop rows without country or year
df.dropna(subset=['country', 'year'], inplace=True)
```
